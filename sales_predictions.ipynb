```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print('libs loaded')
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
```

```python
sales = pd.read_csv('datas/sales_predictions.csv');
sales.describe()
```

```python
sales.info()
```

```python
sales.describe()
```

```python
sales = sales.sample(650, replace=True)
sales['date'] = pd.to_datetime(sales['date'])
sns.relplot(x=sales['date'], y='item_price', data=sales, height=8, aspect=3, kind='line')
```

```python
#2   graphique qui va vous permettre de voir lvolution des prix par rapport au temps.
tips = sns.load_dataset("tips")
g = sns.relplot(data=tips, x=sales['date'], y=sales['item_price'])
g.ax.axline(xy1=(10, 2), slope=.2, color="b", dashes=(5, 2))
```
